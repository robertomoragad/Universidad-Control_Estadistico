```python
#Importamos librerias

import numpy as np # importando numpy
from scipy import stats # importando scipy.stats
import pandas as pd # importando pandas
import math
import matplotlib.pyplot as plt
```

```python
#Una empresa fabrica piezas de metal para dispositivos electrónicos
#Analísis de umbrales: se realiza un estudio para conocer la relación entre la variable espesor y
#la tasa de fallas de la pieza puesta en operación durante 1 mes
#622 y 638
espesor=[610,611,615,620,621,625,628,630,635,637,638,640,642,645]
tasa_fallas=[0.72,0.66,0.65,0.42,0.07,0.04,0.02,0.005,0.05,0.03,0.03,0.05,0.39,0.52]
costo_fallas=[0.3,0.28,0.25,0.27,0.26,0.24,0.25,0.24,0.27,0.24,0.3,0.57,0.59,0.7]

plt.scatter(espesor, tasa_fallas)
plt.plot(espesor, tasa_fallas)

plt.scatter(espesor, costo_fallas)
plt.plot(espesor, costo_fallas)


plt.ylabel("Tasa de falla")
plt.xlabel("espesor pieza (cm)")
plt.title("Relación del espesor con fallas y costos")
plt.legend(['Tasa fallas','Costo de la falla'], bbox_to_anchor=(1.1,1.05))

plt.show()

```

#### Como resultado del análisis anterior, se establece la siguiente especificación ténica para el espesor de la pieza:
630 +- 8 cm

LEI = 622 cm y LES = 638 cm
(LEI: límite de especificación inferior; LES: límite de especificación superior)

```python
#Una empresa fabrica piezas de metal, el espesor de estos productos es una característica 
#Crítica para la Calidad (CTQ).
#Generamos la población de productos que queremos estudiar durante 1 mes 
#la empresa fabrica 100 productos diarios de lunes a viernes (20 días /mes)

mu = 629
sigma = 3.15
size = 2000 #100*20 días

poblacion = np.random.RandomState(seed=1).normal(mu, sigma, size)

#voz cliente
lei = 622
les = 638
```

```python
#Grafiquemos la distribución estadística de la variable "espesor" de la producción mensual

plt.figure(figsize=(10, 5))
plt.hist(poblacion, 50, color='r') #50 clases
plt.xticks(range(615,650,5))
plt.show()
```

```python
#Pero esta empresa no tiene la capacidad de medir pieza por pieza, por tanto aplicará
#un control estadístico de procesos, midiendo 20 piezas de forma aleatoria por día
#generemos la muestra completa de los 20 días muestreados

n=20 #tamaño muestral diario
d=20 #20 días estudiados

muestra20dias = np.random.RandomState(seed=1).choice(poblacion, (n*d))
plt.figure(figsize=(10, 5))
plt.hist(muestra20dias, 50, color='b') #50 clases
plt.xticks(range(615,650,5))
plt.show()
```

```python
#Ahora listemos las muestras de cada día, de forma manual
 
muestra1 = muestra20dias[0:20] #Día 1
muestra2 = muestra20dias[20:40] #Día 2
muestra3 = muestra20dias[40:60] #Día 3
print('muestra 1:',muestra1)
np.mean(muestra1)
```

```python
#Ahora listemos las muestras de cada día de forma automática
t_muestra = 20 #tamaño de cada muestra tomada cada día
n_muestras = 20 #la cantidad de días (del día 1 al día 20)

muestras=[]

for i in range(0, n_muestras):
        muestras.append(muestra20dias[i*t_muestra:(i+1)*t_muestra])

print(muestras)
```

```python
#Ahora calculamos la media muestral (Xbar) y la desviación estandar muestral (S) para cada muestra
#Se aplicará la formula de Desv std insesgada multiplicando por: sqrt(n / n-1)

muestras=[]

for i in range(0, n_muestras):
        muestras.append(muestra20dias[i*t_muestra:(i+1)*t_muestra])

for i in range(0, n_muestras):
    print("muestra", i+1, "Xbar:", muestras[i].mean(), "S:", muestras[i].std()*(math.sqrt(t_muestra/(t_muestra-1))),"n:", len(muestras[i]))
    
```

```python
#Crearemos la lista de Xbarras (medias muestrales) y la lista de S (desv std muestrales)

muestras=[]

for i in range(0, n_muestras):
        muestras.append(muestra20dias[i*t_muestra:(i+1)*t_muestra])

print('Xbarra')
lista_xbarra=[]
for i in range(0, n_muestras):
    lista_xbarra.append(muestras[i].mean())
    print(muestras[i].mean())

print('S')
lista_s=[]
for i in range(0, n_muestras):
    lista_s.append(muestras[i].std()*math.sqrt(t_muestra/(t_muestra-1)))
    print(muestras[i].std()*math.sqrt(t_muestra/(t_muestra-1)))

    
    
#Ahora vamos a inferir la media poblacional y la desv. std poblacional (de corto plazo)
#Media poblacional se infiere con Xbarrabarra (la media de las medias muestrales)
mu_tongo=np.mean(lista_xbarra) #xbarrabarra
print('Mu^=',mu_tongo)

#Desv std. poblacional (de corto plazo) se infiere con Sbarra / c4 (la media de las desv std muestrales / c4 (tabla CEP))
#para n=20 ==> c4=0.9869
c4=0.9869
sbarra=np.mean(lista_s)
sigma_tongo=sbarra/c4
print('Sigma^=',sigma_tongo)


#Ahora vamos a crear los gráficos de control de proceso
#gráfico Xbarra (medias muestrales), para ver la variabilidad entre muestras
#cácular los límites:

x_lcs=mu_tongo + 3*(sigma_tongo/math.sqrt(t_muestra))
x_lc=mu_tongo
x_lci=mu_tongo - 3*(sigma_tongo/math.sqrt(t_muestra))



plt.plot(lista_xbarra)
plt.hlines(y=x_lcs, xmin=0, xmax=t_muestra)
plt.hlines(y=x_lc, xmin=0, xmax=t_muestra)
plt.hlines(y=x_lci, xmin=0, xmax=t_muestra)
plt.ylabel("Xbarra")
plt.xlabel("muestras")
plt.title("Gráfico de control Xbarra")
plt.legend(['Xbarra','Límites de control'],bbox_to_anchor=(1.1,1.05))
plt.show()
```

```python
#Ahora vamos a crear los gráficos de control de proceso
#gráfico S (desv. std muestrales), para ver la variabilidad intra muestras
#cácular los límites:
#buscar B4 y B3 en tablas de control estadístico cuando n=20
b4=1.490
b3=0.510


s_lcs= sigma_tongo * b4
s_lc= sigma_tongo
s_lci= sigma_tongo * b3



plt.plot(lista_s)
plt.hlines(y=s_lcs, xmin=0, xmax=t_muestra)
plt.hlines(y=s_lc, xmin=0, xmax=t_muestra)
plt.hlines(y=s_lci, xmin=0, xmax=t_muestra)
plt.ylabel("S")
plt.xlabel("muestras")
plt.title("Gráfico de control S")
plt.legend(['S','Límites de control'],bbox_to_anchor=(1.1,1.05))
plt.show()
```

Debido a que no se observan puntos por fuera de los límtes de control tanto en gráfico Xbarra como en S, se puede concluir que no existen causas especiales y por lo tanto el proceso se encuentra bajo control estadístico.

```python
#Análisis de capacidad
#cp= voz cliente / voz proceso
cp= (les-lei)/(6*sigma_tongo)
print('CP=',cp)

#cpk= min { (les -mu_tongo)/(3*sigma_tongo)  ; (mu_tongo - lei)/(3*sigma_tongo) }
cpk1=(les - mu_tongo)/(3*sigma_tongo)
cpk2=(mu_tongo - lei)/(3*sigma_tongo)
print('CPk=',min(cpk1,cpk2))

#graficar histrograma
inferir_produccion = np.random.RandomState(seed=1).normal(mu_tongo, sigma_tongo, size)

plt.figure(figsize=(10, 5))
plt.hist(inferir_produccion, 50, color='b') #50 clases
plt.vlines(x=lei, ymin=0, ymax=120, color='g') #LEI
plt.vlines(x=les, ymin=0, ymax=120, color='r') #LES
plt.legend(['LEI','LES'],bbox_to_anchor=(1.1,1.05))

plt.xticks(range(615,650,5))
plt.show()
```
